## Fecha torneo 

```python
    # variable a√±o_torneo
df['tourney_year'] = (df['tourney_date'].astype(str).str.slice(0, 4)).astype(int)
df.head(1)
```

```python
    # variable mes_torneo
df['tourney_month'] = (df['tourney_date'].astype(str).str.slice(4,6)).astype(int)
df.head(1)
```

```python
    #variable estaci√≥n 
estacion=[]
for i in df.tourney_month:
     if i<4:
        i='invierno'
        estacion.append(i)
     elif 4<=i<=6:
        i='primavera'
        estacion.append(i)
     elif 6<i<=9:
        i='verano'
        estacion.append(i)
     else:
        i='oto√±o'
        estacion.append(i)
df['estacion']=estacion
df.head(1)
```

## SCORE 

```python
   #variable sets por partido
sets=[]
for i in df.score:
     sets.append(len(i.split()))
   
df['sets']=sets
df.head(3)
```

```python
     # VARIABLE sets promedio que se hacen por ronda y torneo
df['sets_mean_toronda'] = df.groupby(['tourney_name', 'round'])['sets'].transform('mean')
df[['round']]
```

```python
    # Variable juegos por partido
def suma_juegos_partido(x):
        sets=x.split()
        suma=0
        for set in sets:
            #print(re.sub("\(\d\)","", set))
            set_n=re.sub("\(\d\)","", set)
            score1=re.findall("\d+",set_n)[0]
            score2=re.findall("\d+",set_n)[1]
            suma+=int(score1)
            suma+=int(score2)
        return suma
df['juegos'] = df.score.apply(suma_juegos_partido)
df.head(1)
```

```python
    # variable JUEGOS promedio por ronda y torneo
df['juegos_mean_toronda'] = df.groupby(['tourney_name', 'round'])['juegos'].transform('mean')
df.head(2)
```

```python
    # VARIABLE RATIO DE VICTORIAS (Contando hasta ese partido)

# Inicializar los diccionarios para almacenar victorias y derrotas acumuladas
player_wins = {}
player_losses = {}

# Listas para almacenar los ratios calculados
win_ratios_winner = []
win_ratios_loser = []

# Iterar sobre el DataFrame para calcular los ratios de victorias acumuladas
for index, row in df.iterrows():
        winner = row['winner_id']
        loser = row['loser_id']
               
        # Actualizar victorias y derrotas acumuladas para el ganador
        if winner not in player_wins:
            player_wins[winner] = 0
            player_losses[winner] = 0
        
        # Actualizar el ratio del ganador
        total_games_winner = player_wins[winner] + player_losses.get(winner, 0)
        win_ratio_winner = player_wins[winner] / total_games_winner if total_games_winner > 0 else 0
        win_ratios_winner.append(win_ratio_winner)
        player_wins[winner] += 1
    
        # Actualizar victorias y derrotas acumuladas para el perdedor
        if loser not in player_wins:
            player_wins[loser] = 0
            player_losses[loser] = 0
                
        # Actualizar el ratio del perdedor
        total_games_loser = player_wins[loser] + player_losses[loser]
        win_ratio_loser = player_wins[loser] / total_games_loser if total_games_loser > 0 else 0
        win_ratios_loser.append(win_ratio_loser)
        player_losses[loser] += 1
    
# A√±adir las columnas al DataFrame original
df['winner_win_ratio'] = win_ratios_winner
df['loser_win_ratio'] = win_ratios_loser
df
```
