```python
df.shape
```

```python
df = df.dropna(subset=['minutes'])
df.shape
```

```python
    # Graficar variable minutes

# Crear una figura con subplots
fig, axes = plt.subplots(1, 2,figsize = (15,6))

# 1er Gráfico
sns.histplot(df['minutes'], ax=axes[0], color='aqua',alpha=0.3)
axes[0].set_title('Histograma minutes')

# 2º Gráfico
sns.boxplot(df['minutes'].values,ax=axes[1], color = 'orange')
axes[1].set_title('Boxplot minutes')

# Ajustar diseño
plt.tight_layout()
plt.show()
```

```python
    # gráfica distribución normal target
fig, axes = plt.subplots(1, 2,figsize = (15,6))

# 1er Gráfico
sns.kdeplot(df['minutes'], ax=axes[0], color='cyan')
axes[0].set_title('Minutes sin normalizar')

# 2º Gráfico
sns.kdeplot(np.log(df['minutes']),ax=axes[1], color = 'orange',)
axes[1].set_title('Minutes normalizada')

# Ajustar diseño
plt.tight_layout()
plt.show()
```
