```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import re
from sklearn.impute import SimpleImputer
```

```python
    # Eliminación de columnas sin valor
columnas=['tourney_id','match_num','winner_seed','winner_entry','loser_seed','loser_entry']
df.drop(columns=columnas,inplace= True)
df.head(5)
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
    #quitar filas con minutes nulos
df = df.dropna(subset=['minutes'])
df.shape
```

```python
    # NEXTGEN mecanismo del torneo diferente 
df = df[~df['tourney_name'].str.startswith('NextGen', na=False)]
df.shape
```

```python
   # Eliminar los partidos walkover y retirados
df = df[df['score']!='W/O']
df = df[df['score']!=' W/O']
df = df[df['score']!=' RET']
df.shape
```

```python
    # eliminar partidos de retirados,walkover o descalificados en score
df = df[~df['score'].str.contains('RET', na=False)]
df = df[~df['score'].str.contains('DEF', na=False)]
df = df[~df['score'].str.contains('Def.', na=False)]
df[df['score'].str.contains(' W/O', na=False)]['score']
df.shape
```

```python
    #eliminar partidos con menos sets de los que hay que ganar según best_of
#variable sets por partido
sets=[]
for i in df.score:
     sets.append(len(i.split()))
   
df['sets']=sets
#eliminacion de los que se han jugado 2 y es al mejor de 5
df = df[~((df['sets']==2) & (df['best_of']==5))]
df.shape
```

```python
    # eliminar outliers en minutos>350
df = df.query('minutes<350')
df.shape
```

```python
    # eliminar los partidos de menos de 25 minutos excepto el de shangai que si es real
fila = df[(df['minutes']==25.0) & (df['tourney_name']=='Shanghai')]
df = df[df['minutes']>25.0]
df = pd.concat([df, fila], ignore_index=True)
df.shape
```

```python
 # cambiar las filas con Carpet a partir de 2009 por Hard
df['tourney_date'] = df['tourney_date'].astype(str)
df.loc[(df['surface'] == 'Carpet') & (df['tourney_date'].str.startswith('2009')), 'surface'] = 'Hard'
df['tourney_date'] = df['tourney_date'].astype(int)
```

```python
df.isnull().sum()
```

```python
    # eliminamos los nulos de esas columnas por no poder asignarles valor (visto en numéricas)
df = df.dropna(subset=['winner_rank','winner_rank_points','loser_rank','loser_rank_points'])
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns
```
