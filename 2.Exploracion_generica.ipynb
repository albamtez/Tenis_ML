## Campo de datos:

- **tourney_id** : Identificador único para cada torneo
- **tourney_name**: Nombre del torneo
- **surface**: Tipo de superficie de la pista:
  - *Grass*: Césped
  - *Clay*: Tierra.
  - *Carpet*: Moqueta
  - *Hard*: Pista dura
- **draw_size**: tamaño del cuadro: número de jugadores en él, a menudo redondeado a la potencia de 2 más cercana
- **tourney_level**: nivel del torneo:
  - *A*: Otros torneos del circuito
  - *M* : Master 1000
  - *G*: Grand Slam
  - *F*: Finales del circuito y otros torneos de final de temporada
  - *D*: Copa Davis
- **tourney_date**: Fecha del torneo.Normalmente el lunes de la semana del torneo
- **match_num**: Identificador de partido
- **winner_id**: id del ganador del partido
- **winner_seed**: Número de cabeza de serie.Cuánto mas bajo mejor.Se aplica a la planificación del torneo para que los jugadores de más nivel no se enfrenten entre sí en las primeras rondas 
- **winner_entry**:Valores str de como entró el jugador al torneo: *'Q', 'LL', 'WC', 'PR', 'SE', 'ALT'*
- **winner_name**: nombre del ganador del partido
- **winner_hand**: mano que utiliza el jugador ganador del partido
- **winner_ht**: altura del ganador del partido
- **winner_ioc**: Nacionalidad del ganador del partido
- **winner_age**: Edad del ganador del partido
- **loser_id**: id del perdedor del partido
- **loser_seed**: Número de cabeza de serie.
- **loser_entry**: Valores str de como entró el jugador al torneo: *'Q', 'LL', 'WC', 'PR', 'SE', 'ALT'*
- **loser_name**:nombre del perdedor del partido
- **loser_hand**:mano que utiliza el jugador perdedor del partido
- **loser_ht**: Altura del perdedor del partido
- **loser_ioc**: Nacionalidad del perdedor del partido
- **loser_age**: Edad del perdedor del partido
- **score**: Marcador final del partido
- **best_of**: Si el torneo es al mejor de *3* sets o de *5* sets (2 ganados o 3 ganados) 
- **round**: ronda del torneo 
- **minutes**: Duración del partido en minutos. TARGET
- **w_ace**: Número de aces del ganador
- **w_df**: Número de dobles faltas cometidas por el ganador 
- **w_svpt**: Número de primeros servicios efectuados por el ganador
- **w_1stIn**: Número de primeros servicios válidos del ganador 
- **w_1stWon**: Número de primeros servicios ganados por el ganador
- **w_2ndWon**: Número de segundos servicios ganados por el ganador 
- **w_SvGms**: Número de juegos ganados por el ganador 
- **w_bpSaved**: Número de puntos de break salvados por el ganador 
- **w_bpFaced**: Número de puntos de break afrontados por el ganador 
- **l_ace**: Numero de aces del perdedor 
- **l_df**: Número de dobles faltas cometidas por el perdedor 
- **l_svpt**:Número de primeros servicios efectuados por el perdedor 
- **l_1stIn**: Número de primeros servicios válidos del perdedor  
- **l_1stWon**: Número de primeros servicios ganados por el perdedor
- **l_2ndWon**: Número de segundos servicios ganados por el perdedor 
- **l_SvGms**: Número de juegos ganados por el ganador
- **l_bpSaved**: Número de puntos de break salvados por el perdedor
- **l_bpFaced**: Número de puntos de break afrontados por el perdedor
- **winner_rank**: Número de ranking del ganador 
- **winner_rank_points**: Número de puntos de ranking del ganador
- **loser_rank**: Número de ranking del perdedor 
- **loser_rank_points**: Número de puntos de ranking del perdedor
