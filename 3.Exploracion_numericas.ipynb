```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings(action='ignore',category=FutureWarning)
```

```python
df = pd.read_csv('tenis_bruto.csv')
df
```

```python
    # Quitar filas correspondientes a minutes nulos
df.dropna(subset='minutes',inplace=True)
df.shape
```

```python
df.columns
```

```python
    # quitar variables sin interes en su análisis
df.drop(columns=['tourney_id','match_num','winner_id','loser_id'], inplace = True)
```

```python
    # división en numéricas
df_num = df._get_numeric_data()
df_num.head()
```

```python
df_num.columns
```

```python
df_num.isnull().mean()*100
```

```python
df_num[['winner_ht','winner_age','loser_ht','loser_age']].describe()
```

## Análisis Univariado

```python
    # Graficar variables
for i in df_num.columns:
    # Crear una figura con subplots
     fig, axes = plt.subplots(1, 2,figsize = (15,6))
    
    # 1er Gráfico
     sns.histplot(df[i], ax=axes[0], color='aqua',alpha=0.3)
     axes[0].set_title(f'Histograma {i}')
    
    # # 2º Gráfico
    # sns.scatterplot(df['minutes'],ax=axes[1],fill = True, color = "green", alpha = 0.2)
    # axes[1].set_title('Distribución minutes')
    
    # 3er Gráfico
     sns.boxplot(df[i].values,ax=axes[1], color = 'orange')
     axes[1].set_title(f'Boxplot {i}')
    
    # Ajustar diseño
     plt.tight_layout()
     plt.show()
```

## Análisis Bivariado 

```python
for variable in df_num.columns:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data = df, y=df_num[variable],x=df['minutes'])
        plt.title(f'Scatter de {variable}')
        plt.xlabel('Minutes')
        plt.ylabel(variable)
        plt.show()
```
